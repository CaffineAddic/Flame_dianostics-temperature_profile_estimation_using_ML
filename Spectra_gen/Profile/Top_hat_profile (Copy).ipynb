```python
def bolts_int(A2, A3, x, x_o):
    A1 = 303
    b = A1 + (A2-A1)/(1+np.exp((x-x_o)/A3))
    return b
```

```python
A2 = 538.0080602192687
A3 = 0.5503972556122777
x_o = 28.00498305421301
```
